**ARITHMETIC OPERATORS**

```python
num1=int(input("Enter first number:"))
num2=int(input("Enter second number:"))
```

```python
sum=num1+num2
print("sum=",sum)
```

```python
sub=num1-num2
print("diiference=",sub)
```

```python
mul=num1*num2
print("product=",mul)
```

```python
div=num1/num2
print("division=",div)
```

```python
mod=num1%num2
print("modulus=",mod)
```

```python
pow=num1**num2
print(pow)
```

```python
div=num1/num2
print("division=",div)
d=num1//num2
print("integer dicision=",d)
```

**ASSIGNMENT OPERATOR**

```python
a=5
b=6
sum=a+b
print("sum=",sum)

result=sum   #assigning the value of sum into result
result+=a    # result=result+a
print("result=",result)

result-=a    #result=result-a
print("result=",result)

```

**RELATIONAL OPERATORS**

```python
n1=60
n2=30
print("The result of equals to opearator is",n1==n2) #The == operator checks if the value of two opearnds are equal
```

```python
print("The result of not equals to opearator is",n1!=n2) #The != operator checks if the value of two opearnds are not equal
```

```python
print("The result of greater than opearator is",n1>n2) #The > operator checks if the value of left opearnd is greater than right operand
```

```python
print("The result of greater than or equals opearator is",n1>=n2) #The >= operator checks if the value of left opearnd is greater than or equal to right operand
```

```python
print("The result of less than opearator is",n1<n2) #The < operator checks if the value of left opearnd is less than right operand
```

```python
print("The result of less than or equals opearator is",n1<=n2) #The <= operator checks if the value of left opearnd is less than or equal to right operand
```
